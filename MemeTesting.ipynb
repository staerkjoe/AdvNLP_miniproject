```python
from PIL import Image
import matplotlib.pyplot as plt
import os

image_names = [f"image_{i}.jpg" for i in range(1, 5)] # Assuming .jpg extension for exact match
image_dir = os.path.join(path, "memotion_dataset_7k", "images") # Assuming images are in an 'images' subdirectory

for image_name in image_names:
    image_path = os.path.join(image_dir, image_name)
    try:
        img = Image.open(image_path)
        plt.imshow(img)
        plt.title(image_name)
        plt.axis('off')
        plt.show()
    except FileNotFoundError:
        print(f"Image not found: {image_name}")
    except Exception as e:
        print(f"Could not open image {image_name}: {e}")
```
